```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv('test_out_test.csv')
x = data.index.to_list()
y1 = data['PV1-0.time-based_0-pv_gen_out'].tolist()
y2 = data['PandaController1-0.time-based_0-incoming'].tolist()
for i in range(len(y2)):
    y2[i] = -y2[i]
plt.plot(x,y1,y2,'--')
plt.xlabel('time step')
plt.ylabel('output power [MW]')

plt.legend(['model','element'])
```

```python
y1 = data['FossilCoal1-0.time-based_0-gen_pow'].tolist()
y2 = data['PandaController1-0.time-based_0-outgoing'].tolist()
plt.plot(x,y1,y2,'--')
plt.xlabel('time step')
plt.ylabel('output power [MW]')

plt.legend(['model','element'])
```
